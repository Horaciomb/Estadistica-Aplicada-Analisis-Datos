```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spicy
```

# 1. Wellstone, Inc., produce y comercializa fundas para teléfonos celulares en una variedad de colores.
### A la compañía le gustaría circunscribir sus planes de producción a cinco diferentes colores: blanco brillante, negro metálico, lima magnético, naranja tangerina y rojo fusión. En consecuencia, montó un quiosco  en  el  Mall  of  America  por  varias  horas  y  preguntó,  a  personas elegidas  de  forma  aleatoria,  qué  color  de  funda  era  su  favorito.  Los resultados fueron los siguientes: 

```python
modelo=np.array(['blanco brillante', 'negro metálico','lima magnético', 'naranja tangerina' , 'rojo fusión'])
valor=np.array([130,104,325,455,286])
```

```python
df=pd.DataFrame({
    "Modelo":modelo,
    "Valor":valor
})
df=df.sort_values(by='Valor',ascending=False)
df
```

## a) ¿Que nombre recibe la tabla?

Es una tabla de distribución de frecuencias, pues esta midiendo la frecuencia en la que las personas entrevistadas al azar mencionaron su color funda favorito, y basado en la frecuencia de reperticiones se armó la tabla

## b) Elabora una gráfica de barras para la tabla

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Modelo',y='Valor')
plt.xlabel('Modelo')
plt.ylabel('Valor')
plt.title('Preferencia de colores de fundas para celulares')
plt.xticks(rotation=45)  
plt.show()
```

Acá podemos ver gráficamente que las fundas color naranja, lima y rojo son los colores favoritos de los encuestados, y que los colores blanco y negro son los menos preferidos por los encuestados

## c) Dibuje una gráfica de pastel

```python
plt.figure(figsize=(8,8))
sns.set_theme(style="darkgrid")
plt.pie(df['Valor'], labels=df['Modelo'], autopct='%1.1f%%', startangle=140)
plt.title('Distribución porcentual de la preferencia de colores')
plt.show()
```

En esta gráfica ya podemos ver la proporción de los colores favoritos, donde se nota que la naranja y la lima representan más de la mitdad de la proporción de los colores, siendo naranja el m´sa representativo seguido por lima

## d) Si Wellstone, Inc., tiene planes de producir un millón de fundas para teléfonos celulares, ¿cuántas de cada color debería producir?

```python
total = df['Valor'].sum()
produccion_total = 1000000

df['Produccion'] = (df['Valor'] / total) * produccion_total
df['Produccion'] = df['Produccion'].round(0).astype(int)  
df[['Modelo', 'Produccion']]
```

Con esta tabla ya podemos ver la suegerenicia de cantidad a producir por cada color de funddam, basado en los encuestados iniciales, esto ya representa una tendencia de favoritismos por ciertos colores lo que de mayor probabilidad a que los colores se vendan más

# 2. En una escuela secundaria llamada "Instituto Arquímedes"
### se realizó una evaluación de matemáticas para dos grupos de estudiantes, A y B, con el propósito de analizar y comparar su rendimiento en la materia. El grupo A es conocido por tener un enfoque más tradicional en la enseñanza, mientras que el grupo B se beneficia de métodos educativos más modernos y prácticos. La evaluación constó de 15 problemas matemáticos desafiantes. Cada grupo de estudiantes se sometió a la misma prueba y se registraron las puntuaciones obtenidas por cada estudiante en ambas pruebas. A continuación, se presentan los resultados de las puntuaciones obtenidas por los estudiantes en ambas pruebas:

```python
grupo_a=np.array([78, 82, 83, 66, 88, 73, 89, 71, 96, 80, 90, 73, 79, 85, 88])
grupo_b=np.array([89, 92, 77, 95, 62, 89, 75, 90, 82, 71, 96, 83, 85, 75, 80])
```

Calcula las siguientes medidas estadísticas para los grupos A y B y 
compáralas:

### Media: Calcula la media aritmética de las puntuaciones para cada grupo y compáralas. 
*   ¿Qué grupo obtuvo una puntuación promedio más alta? 

```python
media_a=np.mean(grupo_a)
media_b=np.mean(grupo_b)
print(f"La media del grupo A es: {media_a}")
print(f"La media del grupo B es: {media_b}")
if(media_a>media_b):
    print(f"El grupo A obuvo mayor promedio que grupo B con {media_a:.2f} de nota promedio")
else:
    print(f"El grupo B obuvo mayor promedio que grupo A con {media_b:.2f} de nota promedio")
```

### Mediana: Encuentra la mediana de las puntuaciones para cada grupo. 
*   ¿Cómo se comparan las medianas entre los grupos? 
*   ¿Qué grupo tiene una distribución de puntuaciones más equilibrada? 


```python
mediana_a=np.median(grupo_a)
mediana_b=np.median(grupo_b)
print(f"La mediana del grupo A es: {mediana_a}")
print(f"La mediana del grupo B es: {mediana_b}")
if mediana_a > mediana_b:
    print("El grupo A tiene una mediana más alta.")
elif mediana_a < mediana_b:
    print("El grupo B tiene una mediana más alta.")
else:
    print("Ambos grupos tienen la misma mediana.")
```

```python
desviacion_a = np.std(grupo_a)
desviacion_b = np.std(grupo_b)
print(f"La desviacion del grupo A es: {desviacion_a}")
print(f"La desviacion del grupo B es: {desviacion_b}")
if desviacion_a < desviacion_b:
    print("El grupo A tiene una distribución de puntuaciones más equilibrada.")
else:
    print("El grupo B tiene una distribución de puntuaciones más equilibrada.")
```

Acá con este ejercicio podemos observar que **una desviación menor indica que es más equilibrada** que la otra desviación, indica que los datos varian menos entre sí

### Moda: Identifica si hay alguna moda en las puntuaciones de cada grupo. 
*   ¿Existen puntuaciones que se repiten con mayor frecuencia en alguno de los grupos?

```python
moda_a=spicy.stats.mode(grupo_a)
moda_b=spicy.stats.mode(grupo_b)
print(f"La moda del grupo A es: {moda_a.mode} y se repite {moda_a.count} veces")
print(f"La moda del grupo B es: {moda_b.mode} y se repite {moda_b.count} veces")

if moda_a.count > moda_b.count:
    print("El grupo A tiene una puntuación que se repite con mayor frecuencia.")
elif moda_a.count < moda_b.count:
    print("El grupo B tiene una puntuación que se repite con mayor frecuencia.")
else:
    print(f"Ambos grupos tienen puntuaciones que se repiten con la misma frecuencia: {moda_a.count} veces.")
```

### Desviación Estándar: Calcula la desviación estándar para cada grupo. 
*   ¿Qué grupo tiene una dispersión más amplia en sus puntuaciones? 
*   ¿Qué grupo tiene una concentración más alta de puntuaciones alrededor de la media?  
Basándote en los cálculos de las medidas estadísticas, proporciona una conclusión sobre el rendimiento de los grupos A y B en la prueba de matemáticas. 
*   ¿Puedes inferir si uno de los métodos educativos está teniendo un impacto significativo en el rendimiento de los estudiantes en comparación con el otro?

```python
desviacion_a = np.std(grupo_a)
desviacion_b = np.std(grupo_b)
print(f"La desviación estándar del grupo A es: {desviacion_a:.2f}")
print(f"La desviación estándar del grupo B es: {desviacion_b:.2f}")
```

```python
if desviacion_a > desviacion_b:
    dispersion_amplia = "El grupo A tiene una dispersión más amplia en sus puntuaciones."
    concentracion_alta = "El grupo B tiene una concentración más alta de puntuaciones alrededor de la media."
elif desviacion_a < desviacion_b:
    dispersion_amplia = "El grupo B tiene una dispersión más amplia en sus puntuaciones."
    concentracion_alta = "El grupo A tiene una concentración más alta de puntuaciones alrededor de la media."
else:
    dispersion_amplia = "Ambos grupos tienen la misma dispersión en sus puntuaciones."
    concentracion_alta = "Ambos grupos tienen la misma concentración de puntuaciones alrededor de la media."

print(dispersion_amplia)
print(concentracion_alta)
```

```python
print(f"""
Conclusión sobre el rendimiento de los grupos A y B en la prueba de matemáticas:
- La media de las puntuaciones del grupo A es: {media_a:.2f}
- La media de las puntuaciones del grupo B es: {media_b:.2f}
- La desviación estándar del grupo A es: {desviacion_a:.2f}
- La desviación estándar del grupo B es: {desviacion_b:.2f}
- {dispersion_amplia}
- {concentracion_alta}
""")
```

Basándonos en estos resultados:
*   Podemos inferir que el grupo con la menor desviación estándar tiene una mayor concentración de puntuaciones alrededor de la media,
lo que puede indicar un **rendimiento más consistente** (en este caso **Grupo A**)

*   Si bien el **Grupo B** tiene una media más alta, también tiene una dispersión mayor sobre esa media.
*   Basándonos en estos dos parámetros estadísticos de **media y desviación** podrimos infererir que el **Grupo A** tiene mejor resultado
